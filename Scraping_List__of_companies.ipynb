```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from google.colab import drive
drive.mount('/content/drive')
```

```python
url='https://en.wikipedia.org/wiki/List_of_companies_of_Pakistan'
response=requests.get(url)
soup=BeautifulSoup(response.content,'lxml')
```

```python
table=soup.find('table',class_='wikitable sortable')
```

```python
coloumns=table.find_all('th')
```

```python
new_coloumns=[ elements.text.strip() for elements in coloumns ]
```

```python
new_coloumns
```

```python
df=pd.DataFrame(columns=new_coloumns)
```

```python
df
```

```python
coloumn_row=table.find_all('tr')[1:]
```

```python
rows_data = []

# Iterate through each row
for row in coloumn_row:
    single_row = row.find_all('td')
    # Extract text from each <td> element and strip whitespace
    data = [element.text.strip() for element in single_row]
    # Construct a dictionary with column names as keys and extracted data as values
    row_data = {'Name': data[0], 'Industry': data[1], 'Sector': data[2], 'Headquarters': data[3], 'Founded': data[4], 'Notes': data[5]}
    # Append the dictionary to the list
    rows_data.append(row_data)

# Create DataFrame from the list of dictionaries
df = pd.DataFrame(rows_data)

# Now, df contains the data from coloumn_row
df
```
